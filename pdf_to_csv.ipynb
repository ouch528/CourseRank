```python
def get_pd(round, page):
    ahh = f"round_{round}"
    path2 = "/Users/kohziyang/Documents/hacknroll/" + ahh + ".pdf"
    outputPath2 = "/Users/kohziyang/Documents/hacknroll/" + ahh + ".csv"
    dfs2 = tabula.io.convert_into(path2, pages = page, format= "CSV", output_path= outputPath2)
    df2 = pd.read_csv(outputPath2)

    # df2 = df2.loc[df2['Course Host\rFaculty/School'] == "School of\rComputing"]
    #print(df)


    df2.columns = ['faculty', 'department', 'course_code', "course_title", "course_class", "vacancy", "demand", "success", "reserve","quota", "timetable", "workload", "other"]
    df2 = df2.drop("faculty", axis = 1)
    df2 = df2.drop('department', axis = 1)
    df2 = df2.drop("course_title", axis = 1)
    df2 = df2.drop('reserve', axis = 1)
    df2 = df2.drop("quota", axis = 1)
    df2 = df2.drop('timetable', axis = 1)
    df2 = df2.drop("workload", axis = 1)
    df2 = df2.drop('other', axis = 1)
    
    
    df2 = df2.iloc[1:]
    df2 = df2[df2.vacancy != "-"]
    print(df2)
    
    df2[["demand"]] = df2[["demand"]].apply(pd.to_numeric)
    df2[["vacancy"]] = df2[["vacancy"]].apply(pd.to_numeric)
    # print(df2.describe())
    df2[f"Rd{round}_TF"] = df2["vacancy"] >= df2["demand"]
    df2[f"Rd{round}_rate"] =  df2["demand"] / df2["vacancy"]
    print(df2)
    df2.to_csv(outputPath2)
    return df2

round_0 = get_pd(1, 2)
```

```python
import glob

# List all CSV files with the pattern "round_all*.csv"
csv_files = glob.glob("round_all*.csv")

# Initialize an empty list to store the DataFrames
dfs = []

# Iterate over each CSV file, read it into a DataFrame, and append to the list
for file in csv_files:
    df = pd.read_csv(file, index_col=False)  # Prevent index from being read
    df = df.drop(columns=["vacancy", "demand", "success"])  # Drop unnecessary columns
    dfs.append(df)

# Perform full outer join (merge) on the 'course_code' and 'course_class' columns
merged_df = dfs[0]

for df in dfs[1:]:
    merged_df = pd.merge(merged_df, df, on=['course_code', 'course_class'], how='outer', suffixes=('', '_duplicate'))

# Clean up the duplicate columns, if any
merged_df = merged_df.loc[:, ~merged_df.columns.str.endswith('_duplicate')]

# Drop duplicate rows based on 'course_code' and 'course_class' (if there are duplicates across different files)
merged_df = merged_df.drop_duplicates(subset=['course_code', 'course_class'])
merged_df.to_csv("merged_output2.csv")
```
