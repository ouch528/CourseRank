```python
#user input
mod_class = {"BT3103L1":7, "BT4222L1":6, "GESS1025L1":1, "QF3103L1":2, "CP2201L1":1, "BT4016L1":7}
year = 2
```

```python
#read csv
import numpy as np
import pandas as pd
df = pd.read_csv("testing.csv")
df = df.replace('#DIV/0!', np.inf)
df['ModClass'] = df["Mod"] + df["Class"]
filtered_df = df[df['ModClass'].isin(mod_class.keys())]
filtered_df
```

```python
filtered_df = filtered_df.sort_values(by=["1st oversubscribed"])
filtered_df = filtered_df.copy()
filtered_df["A"] = filtered_df["ModClass"].map(mod_class)
# Adjust the '1st oversubscribed' value to max round (3) if it exceeds it
max_round = 3
filtered_df['Adjusted oversubscribed'] = filtered_df['1st oversubscribed'].apply(lambda x: min(x, max_round))

# Create a secondary sorting column based on the adjusted oversubscription
filtered_df['Sort rate'] = filtered_df.apply(lambda row: row[f'Rd{row["Adjusted oversubscribed"]} rate'], axis=1)
sorted_df = filtered_df.sort_values(by=['1st oversubscribed', 'A', 'Sort rate'], ascending=[True, False, False])
sorted_df
```
