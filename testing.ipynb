```python
#read csv
import numpy as np
import pandas as pd
df = pd.read_csv(r"D:\NUS\Hackathon & Projects\Hack&Roll\courseregRank\merged_output.csv")
df = df.replace('#DIV/0!', np.inf)
df = df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"])
df['codeclass'] = df["course_code"] + df["course_class"]
filtered_df = df[df['codeclass'].isin(mod_class.keys())]
# Create the '1st oversubscribed' column
filtered_df["1st oversubscribed"] = np.where(
    filtered_df["Rd0_TF"].fillna(True) == False, 0, 
    np.where(
        filtered_df["Rd1_TF"].fillna(True) == False, 1,
        np.where(
            filtered_df["Rd2_TF"].fillna(True) == False, 2,  # Treat NaN as False
            np.where(filtered_df["Rd3_TF"].fillna(True) == False, 3, 4)  # Default to 4 if none match
        )
    )
)
filtered_df
```

```python
filtered_df = filtered_df.copy()
filtered_df["A"] = filtered_df["codeclass"].map(mod_class)
# Adjust the '1st oversubscribed' value to max round (3) if it exceeds it
max_round = 3
filtered_df['Adjusted oversubscribed'] = filtered_df['1st oversubscribed'].apply(lambda x: min(x, max_round))

# Create a secondary sorting column based on the adjusted oversubscription
filtered_df['Sort rate'] = filtered_df.apply(lambda row: row[f'Rd{row["Adjusted oversubscribed"]}_rate'], axis=1)
sorted_df = filtered_df.sort_values(by=['1st oversubscribed', 'A', 'Sort rate'], ascending=[True, False, False])
sorted_df
```

```python
import pandas as pd
import glob

# List all CSV files with the pattern "round_all*.csv"
csv_files = glob.glob("round_all*.csv")

# Initialize an empty list to store the DataFrames
dfs = []

# Iterate over each CSV file, read it into a DataFrame, and append to the list
for file in csv_files:
    df = pd.read_csv(file, index_col=False)  # Prevent index from being read
    df = df.drop(columns=["vacancy", "demand", "success"])  # Drop unnecessary columns
    dfs.append(df)

# Perform full outer join (merge) on the 'course_code' and 'course_class' columns
merged_df = dfs[0]

for df in dfs[1:]:
    merged_df = pd.merge(merged_df, df, on=['course_code', 'course_class'], how='outer', suffixes=('', '_duplicate'))

# Clean up the duplicate columns, if any
merged_df = merged_df.loc[:, ~merged_df.columns.str.endswith('_duplicate')]

# Drop duplicate rows based on 'course_code' and 'course_class' (if there are duplicates across different files)
merged_df = merged_df.drop_duplicates(subset=['course_code', 'course_class'])

# Print or save the final merged DataFrame
merged_df

# Optionally save it to a new CSV file
# merged_df.to_csv("merged_output.csv", index=False)

# Filter for a specific course_code

merged_df
```
